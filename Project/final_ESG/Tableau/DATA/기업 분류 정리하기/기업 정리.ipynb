```python
import pandas as pd
import json
```

`DB 불러오기`

```python
sectors_df = pd.read_excel('./kosdaq.xlsx')

sectors_df

del sectors_df['고유번호']
del sectors_df['종목코드']
del sectors_df['주요제품']
del sectors_df['상장일']
del sectors_df['결산월']
del sectors_df['대표자명']
del sectors_df['홈페이지']
del sectors_df['지역']
del sectors_df['사업자등록번호']
del sectors_df['주소']
del sectors_df['업종코드']
del sectors_df['설립일']

sectors_df
```

`유니크한 전체 업종 정리 및 확인하기`

```python
# 유니크한 업종 정리하기
unique_sectors = list(set([sector for sector in sectors_df['업종']]))
print(len(unique_sectors))

# 전체 업종 확인하기  -> 특징을 찾는 용도
for unique_sector in unique_sectors :
    print(unique_sector)
```

`상위 업종 별로 정리한 파일로 기업 분류하기`

```python
with open('제조업.csv', 'r') as 제조업 :
    제조업 = [sector.strip() for sector in 제조업.readlines()]

with open('건설 및 공사업.csv', 'r') as 건설공사 :
    건설및공사업 = [공사.strip() for 공사 in 건설공사.readlines()]

with open('도매 및 소매업.csv', 'r') as 도매업 :
    도매및소매업 = [도매.strip() for 도매 in 도매업.readlines()]

with open('서비스업.csv', 'r') as 서비스업 :
    서비스업 = [sector.strip() for sector in 서비스업.readlines()]

with open('식품 및 음료 제조업.csv', 'r') as 식품제조 :
    식품및음료제조업 = [sector.strip() for sector in 식품제조.readlines()]

with open('운송 및 물류업.csv', 'r') as 물류업 :
    운송및물류업 = [sector.strip() for sector in 물류업.readlines()]

with open('정보기술 및 통신업.csv', 'r') as 정보통신 :
    정보기술및통신업 = [sector.strip() for sector in 정보통신.readlines()]
```

```python
# 제조업 분류하기
search_sector = '광고업'

# 특정 상위 업종의 하위 업종 정리하기

# 하위 업종의 길이 구하기
print(len([unique_sector for unique_sector in unique_sectors if search_sector in unique_sector]))
     
# 하위 업종 이름 정리하기        
for unique_sector in unique_sectors :
    if search_sector in unique_sector :
        print(unique_sector)
```

```python
중간정산 = 제조업 + 건설및공사업 + 도매및소매업 + 서비스업 + 식품및음료제조업 + 운송및물류업 + 정보기술및통신업
len(중간정산)
```

`json 저장 형태`

```python
divide_sec_dict = {'제조업' : {},
 '건설및공사업' : {},
 '도매및소매업' : {},
 '서비스업' : {},
 '식품및음료제조업' : {},
 '운송및물류업' : {},
 '정보기술및통신업' : {}}
```

```python
for i, sector in enumerate(sectors_df['업종']) : 
    if sector in 제조업 :
        divide_sec_dict['제조업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 건설및공사업 :
        divide_sec_dict['건설및공사업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 도매및소매업 :
        divide_sec_dict['도매및소매업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 서비스업 :
        divide_sec_dict['서비스업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 식품및음료제조업 :
        divide_sec_dict['식품및음료제조업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 운송및물류업 :
        divide_sec_dict['운송및물류업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
    elif sector in 정보기술및통신업 :
        divide_sec_dict['정보기술및통신업'][str(sectors_df['회사명'].iloc[i])] = sectors_df['업종'].iloc[i]
```

```python
divide_sec_dict
```

```python
with open('./업종 별 기업 분류.json','w',encoding='UTF-8-sig') as file :
    file.write(json.dumps(divide_sec_dict, ensure_ascii=False))
```

`데이터 프레임 저장형태`

```python
상위업종 = []
for sector in sectors_df['업종'] :
    if sector in 제조업 :
        상위업종.append('제조업')
    elif sector in 건설및공사업 :
        상위업종.append('건설 및 공사업')
    elif sector in 도매및소매업 :
        상위업종.append('도매 및 소매업')
    elif sector in 서비스업 :
        상위업종.append('서비스업')
    elif sector in 식품및음료제조업 :
        상위업종.append('식품 및 음료제조업')
    elif sector in 운송및물류업 :
        상위업종.append('운송 및 물류업')
    elif sector in 정보기술및통신업 :
        상위업종.append('정보기술 및 통신업')
```

```python
sectors_df['상위업종']  = 상위업종
```

```python
sectors_df.rename(columns={'업종' : '하위업종'}, inplace = True)
```
