```python
import pandas as pd
```

```python
# 우수 재활용(GR) 인증업체 현황 
excel_file_path = './green_datas/green_company_GR.xlsx'
data = pd.read_excel(excel_file_path)

# 첫 번째 행을 칼럼으로 사용하여 데이터프레임을 생성
df_1 = pd.DataFrame(data.values[1:], columns=data.iloc[0])
df_1 = df_1.drop(columns=['GR기준', '규격명', '종류', '유효인증번호', '만료기간', '유효기간', '대표자'])

# '인증일자' 칼럼을 datetime 형식으로 변환합니다.
df_1['인증일자'] = pd.to_datetime(df_1['인증일자'], format='%Y.%m.%d')
df_1['인증일자'] = df_1['인증일자'].dt.strftime('%Y-%m-%d')
df_1
```

```python
# 저탄소 인증제품 현황
excel_file_path = './green_datas/green_company_저탄소.xlsx'
data = pd.read_excel(excel_file_path)

# 첫 번째 행을 칼럼으로 사용하여 데이터프레임을 생성
df_2 = pd.DataFrame(data.values[4:], columns=data.iloc[3])
df_2 = df_2.drop(columns=['인증번호', '종료', '인증제품명'])

df_2 = df_2.drop_duplicates(subset=['기업명'])
df_2['연번'] = range(1, len(df_2) + 1)
df_2
```
